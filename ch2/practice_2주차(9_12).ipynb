```python
import pandas as pd

# 예시 데이터
array_dict = {'mag': [1, 2, 3, 4, 5]}

# 'mag' 키의 값을 Series로 변환
mag_series = pd.Series(array_dict['mag'])

# Series 객체의 속성 확인
print(mag_series)
print(mag_series.index)  # 인덱스
print(mag_series.values) # 값
print(mag_series.dtype)  # 데이터 타입
```
